```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(14,9)})
```

```python
data['Pawpularity'].hist(bins=50)
plt.axvline(data.Pawpularity.mean(), color='r', linestyle='--', label="Mean")
plt.axvline(data.Pawpularity.median(), color='g', linestyle='-', label="Median")
plt.axvline(data.Pawpularity.mode().values[0], color='c', linestyle='-', label="Mode")
```

```python
data[data['Pawpularity'] == 100]
```

```python
fig, ax = plt.subplots(12, 3, figsize=(14,40))

for a in ax.ravel():
    a.set(xticks=[], yticks=[])

for r in range(12):
    label = data.columns[r+1] # first one is Id, code is currently a mess :(
    count = data[label].value_counts().sort_values()
    for i in [1, 0]:
        img_id = data[data[label] == i].sample()['Id'].values[0]
        img = plt.imread(f'../input/petfinder-pawpularity-score/train/{img_id}.jpg')
        c = 0 if i == 1 else 2
        ax[r, c].imshow(img)
        ax[r, c].set_title(f'{label}={i}')
    ax[r, 1].pie(count, labels=[1, 0], autopct='%1.1f%%', wedgeprops = {'linewidth': 3})
    ax[r, 1].set_title(f'{label}', fontweight='bold', fontsize=20)

fig.tight_layout()
fig.show()
```
